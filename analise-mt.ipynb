```python
microdados.dropna(subset=['TX_RESPOSTAS_MT'], inplace=True)
microdados["TX_RESPOSTAS_MT"]
```

```python
# selecionei apenas as respostas da prova de cor amarela, código == 1076
resp_MT = microdados["TX_RESPOSTAS_MT"].loc[microdados["CO_PROVA_MT"] == 1076]
if len(resp_MT) > 0:
    primeiro_elemento = resp_MT.iloc[0]
    print(primeiro_elemento)
else:
    print("Não há valores na série filtrada.")

```

```python
#verificando se as 45 respostas estão aqui. Tudo ok!
len(resp_MT.iloc[0])
```

```python

# Converter a coluna em uma lista de strings, onde cada string contém as 45 respostas de uma prova
lista_respostas = resp_MT.tolist()

# Criar lista com as respostas para cada questão
lista_questoes = []
for i in range(45):
    lista_questoes.append([resp[i] for resp in lista_respostas])

# Criar dataframe a partir da lista e atribuir nomes de coluna apropriados
df_questoes = pd.DataFrame(lista_questoes).T
df_questoes.columns = ["Q" + str(i+1) for i in range(45)]
df_questoes
```

```python

# Dicionário de mapeamento para os novos nomes das colunas
mapeamento = {}
for i in range(1, 46):
    mapeamento[f'Q{i}'] = f'Q{i+135}'

# Renomear as colunas do DataFrame
df_questoes = df_questoes.rename(columns=mapeamento)

# Exibir o DataFrame com as colunas renomeadas
print(df_questoes)
```

```python

# Dicionário para armazenar as contagens das colunas
contagens = {}

# Loop pelas colunas do DataFrame
for coluna in df_questoes.columns:
    counts = df_questoes[coluna].value_counts()
    contagens[coluna] = counts

# Criação do DataFrame com as contagens
df_contagens = pd.DataFrame(contagens)

# Exibir o DataFrame com as contagens
print(df_contagens)
```
